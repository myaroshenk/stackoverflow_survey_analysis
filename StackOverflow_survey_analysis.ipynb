<strong>Завдання 3.</strong> Статистичний аналіз досвіду респондентів

Порахуємо кількість рядків з пропущеним WorkExp:

```python
df_raw['WorkExp'].isna().sum()
```

Перевіримо міри центральної тенденції датафрейму, включно з NaN-значеннями:

```python
print(f'Mean: {df_raw['WorkExp'].mean().round()}')
print(f'Median: {df_raw['WorkExp'].median()}')
print(f'Mode: {df_raw['WorkExp'].mode()[0]}')
```

Із розрахунків видно, що median < mean, тому для заповнення пустих значень, без значних спотворень результатів, візьмемо медіанне значення WorkExp.

Створимо новий датафрейм та заповнимо пропущені значення:

```python
df = df_raw.copy()
df['WorkExp'] = df['WorkExp'].fillna(value = df['WorkExp'].median())
```

Виводимо міри центральної тенденції - середнє, медіану та моду:

```python
print(f'Mean: {df['WorkExp'].mean().round()}')
print(f'Median: {df['WorkExp'].median()}')
print(f'Mode: {df['WorkExp'].mode()[0]}')
```

<strong>Завдання 9.</strong> Аналіз популярності Python по віковим категоріям

В датафреймі python_stat створюємо стовпець WorkedWithPython, який містить значення True/False:

```python
python_stat = df[['ResponseId', 'Age', 'LanguageHaveWorkedWith']].dropna()
python_stat = python_stat.copy()
python_stat['WorkedWithPython'] = df['LanguageHaveWorkedWith'].str.contains('Python', case = False, na = False)
python_stat.head(5)
```

Перевіряємо, чи правильно виконано фільтрацію

```python
python_stat.isna().sum()
```

Створюємо датафрейм python_by_age, який містить інформацію про кількість респондентів, що програмують на Python, сумарну кількість респондентів та процентне співвідношення у розрізі вікових категорій:

```python
python_by_age = python_stat.groupby('Age').agg(WorkedWithPython = ('WorkedWithPython', 'sum'), Total = ('ResponseId', 'count')).reset_index()
python_by_age['PythonPercentage'] = round((python_by_age['WorkedWithPython'] / python_by_age['Total']) * 100)
python_by_age
```

Відображаємо лише необхідні стовпці:

```python
python_by_age[['Age', 'PythonPercentage']]
```

<strong>Завдання 10.</strong> Аналіз індустрій серед високооплачуваних віддалених працівників

Створюємо маску для фільтрації респондентів, що працюють віддалено:

```python
remote_mask = df['RemoteWork'] == 'Remote'
remote_mask
```

Фільтруємо респондентів та обираємо необхідні стовпці:

```python
remote_workers_stat = df.loc[remote_mask, ['ResponseId', 'ConvertedCompYearly', 'Industry']].dropna().reset_index(drop = True).sort_values(by = 'ConvertedCompYearly',ascending = False)
remote_workers_stat
```

Знаходимо значення Q3 для річної компенсації (ConvertedCompYearly):

```python
remote_workers_quant_75 = remote_workers_stat['ConvertedCompYearly'].quantile(0.75)
remote_workers_quant_75
```

Фільтруємо рядки зі значенням ConvertedCompYearly більше за Q3:

```python
remote_workers_perc_75 = remote_workers_stat[remote_workers_stat['ConvertedCompYearly'] > remote_workers_quant_75].reset_index(drop = True)
remote_workers_perc_75
```

Рахуємо кількість респондентів в розрізі індустрій та сортуємо в порядку спадання:

```python
remote_75_by_industry = remote_workers_perc_75.groupby('Industry').agg(RespondentsCount = ('ResponseId', 'count')).sort_values(by = 'RespondentsCount', ascending = False)
remote_75_by_industry
```
